```python
import pandas as pd
import json
import numpy as np
import re
from urllib.request import urlopen
import pickle
```

identify women philosophers in this graph

there is no gender information on wikipedia

```python
# Opening JSON file
f = open('C:/Users/zhenj/Box/Porfolio/Projects/Project Vox/Data/philosopher_influence/graph-data.json')
 
# returns JSON object as
# a dictionary
data = json.load(f)
 
# Iterating through nodes

potential_list = []
matches = [" she ", ". She ", " her ", ". Her"]



for node in data['nodes']:
    baseurl = "https://en.wikipedia.org/w/api.php?format=json&origin=*&action=query&prop=extracts&exintro&explaintext&redirects=1&pageids=" + node['id']
    response = urlopen(baseurl).read()
    res = str(response)

    if any(x in res for x in matches):
        
        pot_node = node
        pot_node['description'] = res
        potential_list.append(pot_node)
        # print(pot_node)
```

```python
with open('women_philosopher_pot_list.pkl', 'wb') as f:
    pickle.dump(potential_list, f)
```
