```python
df = pd.read_csv('all_philo_twitter_data.csv')
df
```

```python
df['peer follower (log)'] = np.log2(df['peer follower'])
```

```python


fig = px.scatter(df, x="follower", y="time on Twitter",
	            size="peer follower", color="gender",
                 hover_name="accountname", log_x=True, size_max=27)

fig.update_layout(title='Bubble Chart of Philosophers on Twitter (followers)',
                        title_font_family =  "Times New Roman",

                        paper_bgcolor='#f1ead7',
                        plot_bgcolor='#f1ead7',
                        
                        # margin = dict(t=60, l=60, r=60, b=60),

                        legend_font_family="Times New Roman",
    

                        xaxis=dict(
                            title='Total followers on Twitter',
                            title_font_color = "rgba(40, 60, 130, 1)",
                            title_font_family =  "Times New Roman",
                            gridcolor="rgba(37, 69, 184, 0.2)",
                            type='log',
                            gridwidth=1,
                        ),
                        yaxis=dict(
                            title='Time on Twitter (years)',
                            title_font_family =  "Times New Roman",
                            gridcolor="rgba(77, 106, 209, 0.2)",
                            gridwidth=1,
                        )
)

fig.show()
```

```python
# save the chart as html
fig.write_html("index.html")
```
